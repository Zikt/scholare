## 3. Define your Research Configuration
Instead of creating a JSON file, we just define a standard Python dictionary.

```python
from scholare.config import load_config
from scholare.pipeline import run_pipeline

my_research = {
    "query": "large language models AND agentic reasoning",
    "limit": 10,  # Papers per source (OpenAlex, arXiv, etc) 
    "output_dir": "./notebook_output",
    "categories": {
        "Reasoning": ["chain-of-thought", "cot", "reflection"],
        "Tool Use": ["api", "tool parameter", "executable"]
    },
    "default_category": "General LLMs",
    "download_pdfs": False, # Change to True if you want to download PDFs to the Colab instance
    "sources": ["openalex", "arxiv"] 
}
```

## 4. Run the Pipeline! 
Watch the logs as Scholare automatically searches, deduplicates, and generates your final markdown notes and visualizations.

```python
# Load the dictionary into the Scholare engine
config_obj = load_config(my_research)

# Execute the pipeline
run_pipeline(config_obj)
```

## 5. View Your Results
Everything is saved to the `notebook_output` folder. You can use standard Python to read your CSV or view your Markdown notes.

```python
import pandas as pd
import glob

# Find the newest auto-named folder
output_folders = glob.glob("notebook_output/*/")
latest_run = max(output_folders, key=os.path.getctime)

# Load and display the CSV
df = pd.read_csv(f"{latest_run}/results.csv")
df.head()
```
